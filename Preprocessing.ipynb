```python
import scipy.io as scio
import numpy as np    
import matplotlib.pyplot as plt
import sys
import os
import math
import re
import cv2
from scipy.misc import imsave
from numpy import random
%matplotlib inline
```

```python
def atoi(text) : 
    return int(text) if text.isdigit() else text
```

```python
def natural_keys(text) :
    return [atoi(c) for c in re.split('(\d+)', text)]
```

```python
root_path = ""
melanoma_filenames = []
others_filenames = []
groundtruth = []

for root, dirnames, melanoma_filenames in os.walk("Dataset/melanoma/"):
    melanoma_filenames.sort(key = natural_keys)
    rootpath = root
    
for root, dirnames, others_filenames in os.walk("Dataset/others/"):
    others_filenames.sort(key = natural_keys)
    rootpath = root
    
for root, dirnames, groundtruth in os.walk("Dataset/gt/"):
    groundtruth.sort(key = natural_keys)
    rootpath = root
```

```python
def read_image(path):
    image = cv2.imread(path)
    return image
```

```python
def resize_image(image):
    image = cv2.resize(image,(256,192))
    return image
```

```python
for filename in melanoma_filenames:
    image = read_image('Dataset/melanoma/'+filename)
    image = resize_image(image)
    plt.imsave('Dataset/resized_melanoma/'+filename,image)
```

```python
for filename in others_filenames:
    image = read_image('Dataset/others/'+filename)
    image = resize_image(image)
    plt.imsave('Dataset/resized_others/'+filename,image)
```

```python
for filename in groundtruth:
    image = read_image('Dataset/gt/'+filename)
    image = resize_image(image)
    plt.imsave('Dataset/resized_gt/'+filename,image)
```
